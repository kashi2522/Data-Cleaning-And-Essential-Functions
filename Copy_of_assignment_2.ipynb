**Ans-**

**e. you want to store a static date (today's date) in a cell without it changing every day. what keyboard shortcut should we use??**

**Ans-**

**Ctrl + ; (Control + Semicolon)**

**-This inserts today’s date as a fixed value, so it will not update when the worksheet is reopened.**
